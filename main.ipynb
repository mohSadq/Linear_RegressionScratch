### Forward pass :

```python
def forward(x1,x2,s1,s2,cept):  
    pred = []
    for i in range(n):
        pred.append(s1*x1[i] + s2*x2[i] + cept )
    return pred
```

### MSE :

```python
def cost(pred,real):
    error = []
    for i in range(n):
        error.append( (pred[i] - real[i]) **2 )
    return sum(error)/n
```

### Backpropagation:

```python
def backprop(x1,x2,s1,s2,cept,real):
    ds1  =[]
    ds2  =[]
    dcept=[]
    for i in range (n):
        ds1.append( (s1*x1[i] + s2*x2[i] + cept - real[i]) * x1[i] ) 
        ds2.append( (s1*x1[i] + s2*x2[i] + cept - real[i]) * x2[i] ) 
        dcept.append( s1*x1[i] + s2*x2[i] + cept - real[i]) 
    return sum(ds1)*2/n , sum(ds2)*2/n , sum(dcept)*2/n
```

```python
X1 = [1, 2, 3, 4, 5, 6, 7, 8] # * m1

X2 = [2, 1, 0, 1, 2, 3, 4, 5] # * m2

y = [11, 9, 7, 12, 17, 22, 27, 32]

m1 = 6
m2 = -4
b = 10
n = len(X1)
```

### Gradient descent:

```python
for repeat in range(500):
    y_pred = forward(X1,X2,m1,m2,b)
    MSE    = cost(y_pred,y)
    dm1 , dm2, db = backprop(X1,X2,m1,m2,b,y)
    m1 = m1 - 0.01 * dm1
    m2 = m2 - 0.01 * dm2
    b  = b  - 0.01 * db
    print(f"m1:{m1} | m2:{m2} | b:{b} | MSE:{MSE}")
```

```python
print(f"best m1:{m1} and best m2:{m2} and best b :{b}")
```
